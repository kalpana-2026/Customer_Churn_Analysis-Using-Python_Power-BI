<h3>Data Cleaning</h3>

```python
# Check missing values in each column

print("Missing Values:\n",df.isnull().sum())
```

```python
# Convert 'TotalCharges' column to numeric (it may contain text/blank values)
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
```

```python
# fill missing values
df.fillna(df.mean(numeric_only=True), inplace=True)
```

```python
# check duplicate rows 
print("Duplicate rows:",df.duplicated().sum())
```

```python
# remove duplicates if any
df = df.drop_duplicates()
```

```python
# Handle incorrect values

# Check unique value in categorical columns
print(df["Contract"].unique())
```

```python
print(df["PaymentMethod"].unique())
```

<h3>Basic Churn Analysis</h3>

```python
# Count how many customers churned and how many stayed
print("Churn count:\n", df['Churn'].value_counts())
```

```python
# Churn Percentage
print("Churn percentage:\n",df["Churn"].value_counts(normalize=True)*100)
```

<h3>Exploratory Data Analysis (EDA)</h3>

```python
# Contract vs Churn
# analyze churn based on contract type
print("Contract vs Churn:\n", df.groupby("Contract")["Churn"].value_counts(normalize=True))
```

```python
# Tenure vs Churn
# Compare average tenure of churned vs non-churned customers
print("Average tenure:\n",df.groupby("Churn")["tenure"].mean())
```

```python
# Monthly Charges vs Churn
# Compare average monthly charges
print("Average Monthly Charges:\n",
      df.groupby('Churn')['MonthlyCharges'].mean())
```

```python
# Payment method vs Churn
# Analyze churn by payment method
print("Payment method vs Churn:\n",
      df.groupby("PaymentMethod")["Churn"].value_counts(normalize=True))
```

<h4>High Risk Customers</h4>

```python
# High charges + low tenure = high risk
# create new column "HighRisk"
df['HighRisk'] = np.where(
    (df['MonthlyCharges'] > 70) & (df['tenure'] < 12),
    'Yes', # High risk
    'No'   # Low risk
)

# Display data with new column
df[['MonthlyCharges','tenure','HighRisk']].head()
```

<h4>Long Term Customers</h4>

```python
df["LongTermCust"]= np.where(
    df['tenure']>24,
    "Yes","No"
)

df[['tenure','LongTermCust']].head()
```

<h4>Customer Value</h4>

```python
# Customer Value based on Total Charges
df['CustomerValue'] = np.where(
    df['TotalCharges'] > df['TotalCharges'].median(),
    'High', 'Low'
)

df[['TotalCharges','CustomerValue']].head()
```

```python
# Create tenure groups
df['tenure_group'] = pd.cut(
    df['tenure'],
    bins=[0,12,24,36,48,60,72],
    labels=['0-1 yr','1-2 yr','2-3 yr','3-4 yr','4-5 yr','5-6 yr']
)

df[['tenure','tenure_group']].head()
```
